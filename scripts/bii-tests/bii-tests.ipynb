```python

import os
import requests
import json

url = "https://dev.biii.eurobioimaging.eu/3d-slicer/"
response = requests.get(url)
assert response.status_code == 200
```

```python
software_list_url = "https://dev.biii.eurobioimaging.eu/soft/?_format=json"
response = requests.get(software_list_url)
assert response.status_code == 200
software_list = response.json()
assert len(software_list) > 0
```

```python
for software in software_list:
    if "slicer" in software["title"].lower():
        print(f"Found software: {software['title']}")
        print(json.dumps(software, indent=2))
```

```python
tool_url = "https://dev.biii.eurobioimaging.eu/node/98/?_format=json"
response = requests.get(tool_url)
assert response.status_code == 200
slicer_data = response.json()
assert "open source" in slicer_data["field_has_license"][0]["value"].lower()
```
